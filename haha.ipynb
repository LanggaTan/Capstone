```python
url = "https://www.panganku.org/id-ID/semua_nutrisi"
#payload = {"tipe":"Mentah (Raw)"}
# Send a POST request to the website
response = requests.get(url)
# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.text, 'html.parser')
    # Find all quote elements
    quotes = soup.find_all('td')
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```

```python
cleaned_quotes = [quote.get_text(strip=True) for quote in quotes]
```

```python
id_item = []
name_item = []
i = 1
n = 2
while True:
    try:
        id_item.append(cleaned_quotes[i])
        name_item.append(cleaned_quotes[n])
        i = i+5
        n = n+5
    except Exception as e:
        print(f"An error occurred: {e}")
        break  # Break the loop on any other exception
```

```python
iteration = 0
food_dict = {}
url2 = "https://www.panganku.org/id-ID/view"
for i in id_item:
    payload2 = {'haha': i}
    response2 = requests.post(url2, data=payload2)
    if response2.status_code == 200:
        # Parse the HTML content
        soup2 = BeautifulSoup(response2.text, "html.parser")
        # Find all quote elements
        quotes2 = soup2.find_all("td")
        #Cleaning
        cleaned_quotes2 = [quote.get_text(strip=True) for quote in quotes2]
        cleaned_quotes2 = cleaned_quotes2[21:]
        cleaned_quotes21 = [item.replace(':\xa0', '').strip() for item in cleaned_quotes2]
        #Create Dict
        nutrients_dict = {cleaned_quotes21[i]: cleaned_quotes21[i + 1] for i in range(0, len(cleaned_quotes21), 2)}
        food_dict[name_item[iteration]] = nutrients_dict
        iteration = iteration + 1
        print(f"Progress: {iteration}/1146")
    else:
        print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
        break
```
